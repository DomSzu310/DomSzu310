### Zadanie 1

Przypomnij sobie zadanie o hiperkulach (zadanie 5 z listy 4).

Zilustruj objętości $n$-wymiarowych hiperkul o promieniu $r=1$ na wykresie. Pamiętaj, aby był to wykres punktowy (określony tylko dla liczb naturalnych - sprawdź w dokumentacji lub galerii, jakich opcji funkcji `plot` potrzebujesz). Punkt odpowiadający wartości maksymalnej zaznacz innym kolorem niż pozostałe. 

Kod tworzący rysunek umieść w komórce poniżej.

```python
# z zadania 5 z listy 4 wiemy, że im większa liczba wymiarów, tym mniejsza objętość hiperkuli
# objętość dąży więc ona do 0, zatem możemy wziąć zakres od 0 do 50
# rozpatrujemy hiperkule tylko o r=1, więc możemy to wpisać we wzorze

n = np.linspace(0, 51, 100) # bierzemy równomiernie co 1 liczby z przedziału od 0 do 50
V = np.pi**(n/2)/np.vectorize(np.math.gamma)(n/2 + 1) * 1**n
maksymalna = np.argmax(V) # funkcja zwraca wartość maksymalną

plt.figure(figsize = (20, 6)) # kolejność jest istotna
plt.scatter(n, V, label = 'V(n)')
plt.xticks(np.arange(0, 51, 1))
plt.yticks(np.arange(0, 6, 0.5))
plt.xlim((0, 50))
plt.ylim((0, 5.5))
plt.xlabel('liczba wymiarów (n)')
plt.ylabel('objętość hiperkuli (V)')
plt.title('V(n)')

plt.scatter(n[maksymalna], V[maksymalna], color = 'red', label = 'max V(n)') # zaznaczamy najwięszką wartość innym kolorem
plt.legend(loc = 'upper right')
plt.show()
```

### Zadanie 4

Dla ustalonych $\alpha>0$ i $\beta>0$ definiujemy funkcję $f$ na dodatniej półprostej $(0,+\infty)$ jako
$$f(x)=\frac{\beta^{\alpha}}{\Gamma(\alpha)}x^{\alpha-1}e^{-\beta x},$$
gdzie $\Gamma(\alpha)=\int_{0}^{+\infty}t^{\alpha-1}e^{-t}\mathrm{d}t$ jest wartością tak zwanej funkcji gamma Eulera. Funkcja gamma zaimplementowana jest w module `math` jako [gamma](https://docs.python.org/3/library/math.html#math.gamma), nie jest ona jednak zaimplementowana w bibliotece NumPy.

Na jednym rysunku narysuj wykresy funkcji $f$ dla podanych w tabeli wartości parametrów $\alpha$ i $\beta$.

| Kolor | $\alpha$ | $\beta$  |
|-------|----------|----------|
|<a style='color: red;'> czerwony </a>|<a style='color: red;'> 0.25 </a>|<a style='color: red;'> 0.25 </a> |
|<a style='color: blue;'> niebieski </a> |<a style='color: blue;'> 0.5 </a>|<a style='color: blue;'> 2 </a> |
|<a style='color: green;'> zielony </a> |<a style='color: green;'> 0.75 </a>|<a style='color: green;'> 5 </a> |
|<a style='color: purple;'> fioletowy </a> |<a style='color: purple;'> 2 </a>|<a style='color: purple;'> 2 </a> |
|<a style='color: teal;'> morski </a> |<a style='color: teal;'> 3 </a>|<a style='color: teal;'> 2 </a> |
|<a style='color: orange;'> pomarańczowy </a>|<a style='color: orange;'> 5 </a>|<a style='color: orange;'> 5 </a> |

Wymagania dotyczące wykresów:

1. Kolory wykresów mają się zgadać z podanymi w tabeli;
2. Osie powinny być <b>podpisane</b> (symbolami $x$ i $y$); 
3. <b>Zakres wartości</b> na osi $Ox$ to $[-0.2,4]$, a zakres wartości na osi $Oy$ to $[0,3]$;
4. Należy na rysunku umieścić <b>legendę</b> zawierającą etykiety wszystkich narysowanych wykresów, umieszczoną w prawym górnym rogu;
5. Na rysunku należy zaznaczyć oś $x=0$ czarną linią przerywaną;
6. Na rysunku należy dodać siatkę;
7. Rysunek powinien mieć tytuł "Wykresy funkcji $f$".

Rozwiązanie umieść w komórce poniżej.

```python
# f(x) = x**(a - 1) * np.exp(-b * x) * ((b**a)/np.vectorize(math.gamma)(a))

# bierzemy x0 = 0.0000000000000000000000000000000001, ponieważ jest to wartość dążąca do zera i pozwala uniknąć dzielenia przez 0
# ponadto w treści zadania mamy informację, że definiujemy funkcję na dodatniej półprostej (0, +nieskończoność), co oznacza że 0 nie należy do dziedziny
x = np.linspace(0.0000000000000000000000000000000001, 4.1, 100) # podajemy dziedzinę funkcji

# wykres czerwony
c = x**(0.25 - 1) * np.exp(-0.25 * x) * ((0.25**0.25)/np.vectorize(math.gamma)(0.25))
plt.plot(x, c, color='red', label='f dla α = 0.25, β = 0.25')

# wykres niebieski
n = x**(0.5 - 1) * np.exp(-2 * x) * ((2**0.5)/np.vectorize(math.gamma)(0.5))
plt.plot(x, n, color='blue', label='f dla α = 0.5, β = 2')

# wykres zielony
z = x**(0.75 - 1) * np.exp(-5 * x) * ((5**0.75)/np.vectorize(math.gamma)(0.75))
plt.plot(x, z, color='green', label='f dla α = 0.75, β = 5')

# wykres fioletowy
f = x**(2 - 1) * np.exp(-2 * x) * ((2**2)/np.vectorize(math.gamma)(2))
plt.plot(x, f, color='m', label='f dla α = 2, β = 2')

# wykres morski
m = x**(3 - 1) * np.exp(-2 * x) * ((2**3)/np.vectorize(math.gamma)(3))
plt.plot(x, m, color='c', label='f dla α = 3, β = 2')

# wykres pomarańczowy
p = x**(5 - 1) * np.exp(-5 * x) * ((5**5)/np.vectorize(math.gamma)(5))
plt.plot(x, p, color='orange', label='f dla α = 5, β = 5')

# prosta x=0
plt.axvline(0, color='black', ls='--', label='x = 0')

plt.xlabel('x')
plt.ylabel('y')
plt.xlim((-0.2, 4))
plt.ylim((0, 3))
plt.legend(loc = 'upper right')
plt.grid()
plt.title('Wykresy funkcji f')
plt.show()
```

### Zadanie 5

Zapoznaj się z informacją o mapach kolorów w [dokumentacji Matplotlib](https://matplotlib.org/stable/tutorials/colors/colormaps.html) oraz z kodem poniżej. Sprawdź dokumentację wszystkich funkcji z biblioteki Matplotlib użytych poniżej i zobacz, co oznaczają ich argumenty.

```python
M = np.zeros((100, 100), 'u1') # macierz o wymiarze 100x100

for i in range(100): # i - pionowa oś
    for j in range(100): # j - pozioma oś
        if 48**2 <= (i-50)**2 + (j-50)**2 <= 50**2:  # 48 - grubość fioletowego koła wewnątrz, 50 grubość fioletowego koła na zewnątrz, czyli jeśli wartość z i oraz j jest pomiędzy tymi dwoma to rysowane jest fioletowe koło
            M[i,j] = 1 # zmiana koloru 1-fioletowy
        elif 8**2 <= (i-30)**2 + (j-30)**2 <= 10**2: # modyfikacja lewego oka, taka sama zasada jak wyżej
            M[i,j] = 3
        elif 8**2 <= (i-30)**2 + (j-70)**2 <= 10**2: # modyfikacja prawego oka
            M[i,j] = 3
        elif max(j-10, 90-j) <= i <= 55: # modyfikacja nosa
            M[i, j] = 2
        elif 33**2 <= (i-50)**2 + (j-50)**2 <= 35**2 and i >= 60: # modyfikacja ust, kolejno tak samo jak wyżej, długość uśmiechu
            M[i,j] = 2
            
plt.matshow(M) # 'drukuje' obrazek
plt.set_cmap('magma') # zmiana bazy kolorystycznej
plt.axis('off') # podziałka na osiach
plt.show()
```
